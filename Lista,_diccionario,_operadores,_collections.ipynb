```python

# Listas de datos
compras = [
    "Luis", "Ana", "Luis", "Carlos", "Marta", "Ana", "Sofía",
    "Elena", "Luis", "Carlos", "Pedro", "Juan", "Ernesto",
    "Sofía", "Marcelo", "Francisco", "Pedro", "Juan"
]

registrados = [
    "Ana", "Carlos", "Marta", "Elena"
]

# 1. Clientes nuevos no registrados
nuevos = set(compras) - set(registrados)
print("\n Clientes nuevos no registrados ")
print(nuevos)

# 2. Lista sin duplicados, manteniendo el orden
unicos = list(dict.fromkeys(compras))
print("\nLista de clientes únicos ")
print(unicos)

# 3. Conteo de compras por cliente
from collections import Counter
conteo = Counter(compras)
print("\nConteo de compras por cliente ")
print(conteo)

# 4. Resumen personalizado (solo clientes con más de 1 compra)
resumen = {cliente: f"Ha comprado {veces} veces"
           for cliente, veces in conteo.items() if veces > 1}

print("\n Resumen por cliente frecuente ")
for cliente, mensaje in resumen.items():
    print(cliente, ":", mensaje)
```
